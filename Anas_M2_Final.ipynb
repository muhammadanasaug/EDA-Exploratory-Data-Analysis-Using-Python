## Feature Encoding And Descretization

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['OFFENSE_CODE_GROUP']=encoder.fit_transform(df['OFFENSE_CODE_GROUP'])   
df['OFFENSE_CODE_GROUP'].unique()
```

```python
df['DAY_OF_WEEK']=encoder.fit_transform(df['DAY_OF_WEEK'])   
df['DAY_OF_WEEK'].unique()
```

```python
df
```

```python
dfDummies = pd.get_dummies(df['UCR_PART'], prefix = 'UCR_PART')
```

```python
dfDummies
```

```python
df_new=pd.concat([df, dfDummies], axis=1)
```

```python
df_new
```

```python
df_new['HOUR']  = np.where((df['HOUR'] <=8) ,'Early Morning',
                               np.where((df['HOUR'] >8) & (df['HOUR'] <=12),'Morning',
                               np.where((df['HOUR']>12) & (df['HOUR'] <=16),'Afternoon',
                               np.where((df['HOUR']>16) & (df['HOUR'] <=19),'Evening',
                                       'Night'))))
```
